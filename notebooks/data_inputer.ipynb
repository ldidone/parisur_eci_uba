```python
for input_path, invoice_number, result in saved_errors:
    total_charged = 0
    flag = False
    for res in result:
        expresion_regular = r"[-$\d,.]+"
        salida = "".join(re.findall(expresion_regular, res[1]))
        if (("," in salida or "." in salida) and 
            ("date" not in res[1].lower()) and ("mountain" not in res[1].lower())):
            if len(salida) > 8:
                continue
            if len(salida) > 7:
                salida = salida.replace(",", "")
            else:
                salida = salida.replace(",", ".")
            
            for char, rep in chars:
                salida = salida.replace(char, rep)
            if len(salida) == 1:
                salida = salida.replace(".", "")
            if salida:
                expresion_regular = r"\d+\.\d+(\.\d+)+"
                cont = re.findall(expresion_regular, salida)
                if len(cont) > 0:
                    continue
                    
                expresion_regular = r"\d{4}\.\d{2}-"
                cont_2 = re.findall(expresion_regular, salida)
                if len(cont_2) > 0:
                    continue
                    
                if salida.count('.') > 1:
                    continue           
                
                expresion_regular = r"\d+(?:\.\d+)?-"
                cont_3 = re.findall(expresion_regular, salida)
                if len(cont_3) > 0:
                    continue
                
                salida = float(salida)
                total_charged += salida
                flag = True
    if flag:
        df.loc[df['invoice_number'] == invoice_number, 'total_charged'] = total_charged
```
